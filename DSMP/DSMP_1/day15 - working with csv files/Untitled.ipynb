```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'
}

details = []
last_page = 0

# Set your range here
for p in tqdm(range(1, 5)):
    url = f"https://www.ambitionbox.com/list-of-companies?page={p}"
    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        break

    soup = BeautifulSoup(response.text, 'lxml')
    # Find all company cards first
    cards = soup.find_all('div', class_='companyCardWrapper')

    for card in cards:
        # Basic Info
        name = card.find('h2', class_='companyCardWrapper__companyName').text.strip()
        rating = card.find('div', class_='rating_text').get_text(strip=True)
        desc = card.find('span', class_='companyCardWrapper__interLinking').text.strip()
        logo = card.find('div', class_='companyCardWrapper__companyLogo').img['src']
        
        # Rating Count (e.g., 59.8k)
        rating_count = card.find('span', class_='companyCardWrapper__companyRatingCount').text.strip()
        
        # Tertiary Info (Reviews, Salaries, Jobs, etc.)
        # These are all inside span tags with class 'companyCardWrapper__ActionCount'
        actions = [span.text.strip() for span in card.find_all('span', class_='companyCardWrapper__ActionCount')]
        
        # Safety check: AmbitionBox usually has 6 actions (Reviews, Salaries, Interviews, Jobs, Benefits, Photos)
        reviews = actions[0] if len(actions) > 0 else "0"
        salaries = actions[1] if len(actions) > 1 else "0"
        interviews = actions[2] if len(actions) > 2 else "0"
        jobs = actions[3] if len(actions) > 3 else "0"
        benefits = actions[4] if len(actions) > 4 else "0"
        photos = actions[5] if len(actions) > 5 else "0"

        details.append({
            'Name': name,
            'Rating': rating,
            'Rating_Count': rating_count,
            'Description': desc,
            'Logo_URL': logo,
            'Reviews': reviews,
            'Salaries': salaries,
            'Interviews': interviews,
            'Jobs': jobs,
            'Benefits': benefits,
            'Photos': photos
        })
    
    last_page = p

print(f"Total pages scraped: {last_page}")
print(f"Total items collected: {len(details)}")
```

```python
df = pd.DataFrame(details)
```
