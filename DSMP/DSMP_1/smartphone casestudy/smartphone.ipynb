```python
from bs4 import BeautifulSoup
import pandas as pd

# 1. Load the saved HTML file
with open('smartprix.html', 'r', encoding='utf-8') as f:
    soup = BeautifulSoup(f.read(), 'lxml')

# 2. Find all smartphone containers
product_containers = soup.find_all('div', class_='sm-product')
print(f"Total Smartphones Found: {len(product_containers)}")

all_smartphones = []

for product in product_containers:
    # --- Basic Info ---
    name = product.find('h2').text.strip() if product.find('h2') else "N/A"
    price = product.find('span', class_='price').text.strip() if product.find('span', class_='price') else "N/A"
    
    # --- Rating & Score ---
    score_tag = product.find('div', class_='score')
    spec_score = score_tag.find('b').text.strip() if score_tag else "N/A"
    
    rating_tag = product.find('span', class_='sm-rating')
    rating = rating_tag.get('style').split(':')[-1].replace(';', '').strip() if rating_tag else "N/A"

    # --- Feature Extraction (The List Items) ---
    specs_list = product.find('ul', class_='specs')
    # We grab all <li> tags into a list of strings
    features = [li.text.strip() for li in specs_list.find_all('li')] if specs_list else []
    
    # Standard Smartprix Order:
    # 0: Sim/Connectivity | 1: Processor | 2: RAM/Storage | 3: Battery | 4: Display | 5: Camera | 6: Card | 7: OS
    
    # We use a helper function to avoid "Index out of range" errors
    def get_spec(feat_list, index):
        return feat_list[index] if len(feat_list) > index else "N/A"

    # --- Detailed Mapping ---
    all_smartphones.append({
        'Model': name,
        'Price': price,
        'Rating': rating,
        'Spec Score': spec_score,
        'Sim': get_spec(features, 0),
        'Processor': get_spec(features, 1),
        'RAM': get_spec(features, 2),
        'Battery': get_spec(features, 3),
        'Display': get_spec(features, 4),
        'Camera': get_spec(features, 5),
        'Card': get_spec(features, 6),
        'OS': get_spec(features, 7)
    })

# 3. Save and Clean
df = pd.DataFrame(all_smartphones)

# Remove rows where Model is N/A
df = df[df['Model'] != 'N/A']

# Save to CSV
df.to_csv('smartphones_full.csv', index=False)
print("Complete data saved to smartphones_complete.csv")
len(df)
```
